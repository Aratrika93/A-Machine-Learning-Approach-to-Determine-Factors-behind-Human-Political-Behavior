# Importing files & Exploring the data

```python
import pandas as pd
import numpy as np
```

```python
census_data= pd.read_csv('acs2015_county_data.csv')
census_data.head()
```

```python
census_data.info()
```

```python
voting_data= pd.read_csv('countypres_2000-2016.csv')
voting_data.head()
```

```python
is_2016 =  voting_data['year']==2016
voting_data_2016 = voting_data[is_2016]
voting_data_2016.head()
```

```python
voting_data_2016['year'].value_counts()
```

```python
raj = voting_data_2016['county'].unique()
raj
```

# For the 1st state in the list

```python
sta = voting_data_2016['state'].unique()
sta
```

```python
voting_data_fin = pd.DataFrame()
voting_data_fin
```

```python
is_Alabama =  voting_data_2016['state']=='Alabama'
voting_data_Alabama = voting_data_2016[is_Alabama]

list2 = []
for i in raj:
    df = voting_data_Alabama[voting_data_Alabama['county'] == i]
    max1 = df['candidatevotes'].max()
    df1 = df[df.candidatevotes.eq(max1)]
    df2 = df1[["state","county","party","candidatevotes"]]
    list2.append(df2.T.to_dict())
```

```python
df_val1 = pd.DataFrame()
for i in range(0, len(list2)):
    val = list2[i].values()
    df_val = pd.DataFrame(val)
    df_val1 = df_val1.append(df_val, True)
df_val1
```

```python
list3 = []
for i in range(0,len(df_val1)):
    st1 = df_val1.state[i]
    ct1 = df_val1.county[i]
    party = df_val1.party[i]  
    for j in range(0, len(census_data)):
        st2 = census_data.State[j]
        ct2 = census_data.County[j]
        if ((st1 == st2) & (ct1 == ct2)):
            list3.append(party)
            
```

```python
is_Alabama1 =  census_data['State']=='Alabama'
voting_data_Alabama1 = census_data[is_Alabama1]
voting_data_Alabama1
```

```python
voting_data_Alabama1['Party'] = list3
voting_data_fin = pd.DataFrame(voting_data_Alabama1)
voting_data_fin
```

```python
df_val1 = pd.DataFrame()
for i in range(0, len(list2)):
    val = list2[i].values()
    df_val = pd.DataFrame(val)
    df_val1 = df_val1.append(df_val, True)
df_val1
```

```python
list3 = []
for i in range(0,len(df_val1)):
    st1 = df_val1.state[i]
    ct1 = df_val1.county[i]
    party = df_val1.party[i]  
    for j in range(0, len(census_data)):
        st2 = census_data.State[j]
        ct2 = census_data.County[j]
        if ((st1.lower() == st2.lower()) & (ct1.lower() == ct2.lower())):
            list3.append(party)
```
